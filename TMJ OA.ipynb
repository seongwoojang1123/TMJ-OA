```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 파일 경로 지정
file_path = '/content/drive/MyDrive/20240819 CBCT MRI 데이터 정리.xlsx'

df = pd.read_excel(file_path, sheet_name='정리본')

 #1029행까지와 1030행 이후로 데이터를 나누기
df_part1 = df.iloc[:1029, :]  # 1029행까지

df_part1['SEX (여성-F, 남성-M)'] = df['SEX (여성-F, 남성-M)'].map({'F': 0, 'M': 1})

DF_list1 = df_part1.drop(columns=['Unnamed: 7', '1030행부터 1221 행까지: CBCT와 MRI 모두 있는 환자 '])
df = DF_list1

columns = df.columns
print(len(columns))
print(columns)
df.head()
```

```python
# 파일 경로 지정
file_path = '/content/drive/MyDrive/20240819 CBCT MRI 데이터 정리.xlsx'

df_part2 = pd.read_excel(file_path, sheet_name='정리본', header=1029)

df_part2['SEX'] = df_part2['SEX'].map({'F': 0, 'M': 1})
df2 = df_part2

columns2 = df2.columns
print(len(columns2))
print(columns2)
df2.head()
```

```python
# 'AGE' 열을 숫자로 변환할 수 없는 값들을 NaN으로 처리
df['AGE'] = pd.to_numeric(df['AGE'], errors='coerce')

# NaN 값이 포함된 행을 제거
df = df.dropna(subset=['AGE'])

# 데이터 확인
print(df['AGE'].dtype)  # 'AGE' 열이 숫자로 변환되었는지 확인

```

```python
# CBCT Right TMJ OA 분포
plt.subplot(2, 2, 1)
sns.countplot(x='CBCT_Rt OA (Osteoarthritis) (0=normal, 1=OA)', data=df, palette='Set2')
plt.title('CBCT Right TMJ OA (0=Normal, 1=OA)')
plt.xlabel('CBCT_Rt OA')
plt.ylabel('Count')

# CBCT Left TMJ OA 분포
plt.subplot(2, 2, 2)
sns.countplot(x='CBCT_Lt OA  (Osteoarthritis) (0=normal, 1=OA)', data=df, palette='Set2')
plt.title('CBCT Left TMJ OA (0=Normal, 1=OA)')
plt.xlabel('CBCT_Lt OA')
plt.ylabel('Count')

# MRI Right TMJ OA 분포
plt.subplot(2, 2, 3)
sns.countplot(x='MRI_Rt OA (0=normal, 1=OA)', data=df2, palette='Set2')
plt.title('MRI Right TMJ OA (0=Normal, 1=OA)')
plt.xlabel('MRI_Rt OA')
plt.ylabel('Count')

# MRI Left TMJ OA 분포
plt.subplot(2, 2, 4)
sns.countplot(x='MRI_Lt OA (0=normal, 1=OA)', data=df2, palette='Set2')
plt.title('MRI Left TMJ OA (0=Normal, 1=OA)')
plt.xlabel('MRI_Lt OA')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
#CBCT Right TMJ OA에 대한 레이블 비율 (파이차트)
labels_cbct_Right = df['CBCT_Rt OA (Osteoarthritis) (0=normal, 1=OA)'].value_counts().index
sizes_cbct_Right = df['CBCT_Rt OA (Osteoarthritis) (0=normal, 1=OA)'].value_counts().values

plt.figure(figsize=(15, 5))
plt.pie(sizes_cbct_Right, labels=labels_cbct_Right, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=90)
plt.title('CBCT Right TMJ OA Distribution (0=Normal, 1=OA)')

# CBCT Left TMJ OA에 대한 레이블 비율 (파이차트)
labels_cbct_left = df['CBCT_Lt OA  (Osteoarthritis) (0=normal, 1=OA)'].value_counts().index
sizes_cbct_left = df['CBCT_Lt OA  (Osteoarthritis) (0=normal, 1=OA)'].value_counts().values

plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.pie(sizes_cbct_left, labels=labels_cbct_left, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=90)
plt.title('CBCT Left TMJ OA Distribution (0=Normal, 1=OA)')

# MRI Right TMJ OA에 대한 레이블 비율 (파이차트)
labels_mri_right = df2['MRI_Rt OA (0=normal, 1=OA)'].value_counts().index
sizes_mri_right = df2['MRI_Rt OA (0=normal, 1=OA)'].value_counts().values

# MRI Right TMJ OA 파이 차트
plt.subplot(1, 3, 2)
plt.pie(sizes_mri_right, labels=labels_mri_right, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=90)
plt.title('MRI Right TMJ OA Distribution (0=Normal, 1=OA)')

# MRI Left TMJ OA에 대한 레이블 비율 (파이차트)
labels_mri_left = df2['MRI_Lt OA (0=normal, 1=OA)'].value_counts().index
sizes_mri_left = df2['MRI_Lt OA (0=normal, 1=OA)'].value_counts().values

# MRI Left TMJ OA 파이 차트
plt.subplot(1, 3, 3)
plt.pie(sizes_mri_left, labels=labels_mri_left, autopct='%1.1f%%', colors=['skyblue', 'lightcoral'], startangle=90)
plt.title('MRI Left TMJ OA Distribution (0=Normal, 1=OA)')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (40,25))

for i, column in enumerate(columns[3:]):
    plt.subplot(2,2,i+1)
    plt.title(column.split('\n')[0], size = 15)

    if len(set(df[column])) <=2:
      ratio_0 = (df[column] == 0).sum() / len(df[column])
      ratio_1 = (df[column] == 1).sum() / len(df[column])

      plt.bar([0, 1], [ratio_0, ratio_1], color=['skyblue', 'lightcoral'])
      plt.xticks([0, 1])
      plt.ylim(0, 1)
      plt.ylabel('Ratio', size=15)

    elif len(set(df[column])) > 2:
        # 연속형 데이터의 히스토그램 그리기
        plt.hist(df[column], bins=20, color='skyblue', edgecolor='black')
        plt.ylabel('Frequency', size=15)
        plt.xlabel(column, size=15)
```
